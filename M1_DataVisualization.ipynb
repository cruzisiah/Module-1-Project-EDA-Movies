```python
df1['genres'][0:100].value_counts() #calculation the value count for the top 100 movies in the 'genres' column of df1
```

```python
height = [17,12,10,6,4]
bars = ['Action,Adventure,Sci-Fi', 'Action,Adventure,Fantasy', 'Action,Adventure,Comedy', 'Adventure,Animation,Comedy', 'Action,Crime,Drama']
distribution = np.arange(len(bars))

plt.barh(distribution, height)
plt.yticks(distribution, bars)

plt.gca().invert_yaxis()
plt.title('Top 100 Most Popular Movies by Genre (since 2015)')
plt.xlabel("Count")
plt.ylabel("Genres")

plt.show()

#plotting the distribution of movies across genre sorted by popularity

```

```python
height = [34,22,3,2,2]
bars = ['Documentary', 'Drama', 'Comedy', 'Documentary,Music', 'Comedy,Drama,Family']
distribution = np.arange(len(bars))

plt.barh(distribution, height)
plt.yticks(distribution, bars)

plt.gca().invert_yaxis()
plt.title('Top 100 Highest Rated Movies by Genre (since 2015)')
plt.xlabel("Count")
plt.ylabel("Genres")

plt.show()

#plotting the distribution of movies across genre sorted by rating
```
