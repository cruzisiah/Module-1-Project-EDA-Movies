```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
genres = ['Action', 'Adventure', 'Sci-Fi']
counts = [5,4,3]
ax.bar(genres,counts)
plt.show()
```

```python
df1['genres'][0:100].value_counts()
```

```python
height = [17,12,10,6,4]
bars = ['Action,Adventure,Sci-Fi', 'Action,Adventure,Fantasy', 'Action,Adventure,Comedy', 'Adventure,Animation,Comedy', 'Action,Crime,Drama']
distribution = np.arange(len(bars))

plt.barh(distribution, height)
plt.yticks(distribution, bars)

plt.gca().invert_yaxis()
plt.title('Top 100 Most Popular Movies by Genre (since 2015)')
plt.xlabel("Count")
plt.ylabel("Genres")

plt.show()
```

```python
height = [34,22,3,2,2]
bars = ['Documentary', 'Drama', 'Comedy', 'Documentary,Music', 'Comedy,Drama,Family']
distribution = np.arange(len(bars))

plt.barh(distribution, height)
plt.yticks(distribution, bars)

plt.gca().invert_yaxis()
plt.title('Top 100 Highest Rated Movies by Genre (since 2015)')
plt.xlabel("Count")
plt.ylabel("Genres")

plt.show()
```
