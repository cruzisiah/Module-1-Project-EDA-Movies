# Data Exploration
## Joining IMDB Ratings & IMDB Title Basics

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('imdb.title.basics.cleaned.csv', index_col=0) #import cleaned 'Title Basics' dataset
df2 = pd.read_csv('imdb.ratings.cleaned.csv', index_col=0) #import cleaned 'Ratings' dataset
```

```python
df3 = pd.merge(df1, df2, on='tconst', how='inner') #merge our 2 datasets with 'tconst' as our index
```

```python
df3.sort_values(by=['averagerating','numvotes'], ascending=False)

#sort by 'Average Rating' and 'Numvotes' to detect the correlation between popularity & other factors like runtime
```

```python
df3['numvotes'].median()

#finding the median of the 'numvotes' column to get a sense of what a good critical mass for votes may be.
#of course, a movie with a 100 rating and 1 vote will be judged differently than one with a 100 rating and 1000 votes
```

```python
df3['numvotes'].max()

#similar to our previous code except getting a sense of the movie with the most votes
```

```python
df4 = df3[df3.numvotes >= 49] #removing the rows where numvotes is not equal to the average of the numvotes column
```

```python
df4.sort_values(by=['averagerating','numvotes'], ascending=False)

#re-sort the values now that only movies with a critical mass of votes are included (sorted by averagerating)
```

```python
df4['genres'][0:25].value_counts() #examine which genres have received the highest ratings (sorted by averagerating)
```

```python
df4['genres'].str.split(',', expand=True).head()
```

```python
df5 = df4.sort_values(by=['numvotes','averagerating'], ascending=False)
df5.head()

#re-sort the values now that only movies with a critical mass of votes are included (sorted by numvotes)
```
